```python
import numpy as np
import matplotlib.pyplot as pl
from scipy.interpolate import interp1d

def southpole(z, A = 1.78, B = -0.43, C = -0.0132):
    '''
    Function returns analytical south pole function

    Input:
    z : depth

    output:
    n : refractive index (real)
    '''

    return A+B*np.exp(C*z)
```

```python
def get_eps_profile_from_file(fname):
    prof_data = np.genfromtxt(fname)
    z_prof = prof_data[:,0]
    eps_prof = prof_data[:,1]
    return eps_prof, z_prof

eps_prof, z_prof = get_eps_profile_from_file('NEEM_permittivity.txt')
```

```python
fig = pl.figure(figsize=(5,8),dpi=120)
ax = fig.add_subplot(111)

ax.plot(eps_prof, z_prof, c='b', label='NEEM Eastern Greenland')

ax.grid()
ax.set_ylabel('Depth Z [m]')
ax.set_xlabel('Permittivity')
#ax.set_ylim(30, -10)
ax.set_xlim(1, 3.5)
ax.legend()
fig.savefig('ref_index.png')
ax.set_ylim(200, 0)
pl.show()
```
